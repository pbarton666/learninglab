# Excel versus Python - More "Spreadsheet Functions"

You can do anything with a pandas DataFrame that you can do in a spreadsheet.   In this section we'll explore how to implement more Excel functionality into Pandas objects.


### Cell-wise Built-in Functions

Pandas objects have a lot of functionality "baked in" - in other words there are many methods associated with the objects that can be called directly.

Here are some examples, starting with the sports data set and some operations we may want to perform in a spreadsheet.  Here, we want to "center" the attendance data (make it scale-neurtal):

```python
#Imports
import pandas as pd
from numpy import NaN
import sys
import os

#This is a utility for displaying content side-by-side
#import custom_utils
from custom_utils import  display_wide
from custom_utils.display_wide import  display_wide, plot_image_save

#Some file names
support_dir = "support_files"
img_file_name = os.path.join(support_dir, "sports_normalized.jpg")
csv_file_name = os.path.abspath(os.path.join(support_dir, "sports.csv"))
```

```python
#Read a file
sports = pd.read_csv(csv_file_name)

display_wide([sports, img_file_name], ["Data", "Analysis"],spacing=1)
```

In this example below we apply the built-in DataFrame methods <b>mean()</b> and <b>std()</b>.  You'll note that these are applied to the entire column.  It's as simple as using "dot notation" to name the function and adding opening/closing parens to execute.   

```python
#Add some descriptive statitistics
sports['fan_avg'] = sports['fans'].mean()
sports['fan_stdev'] = sports['fans'].std(ddof=0)
sports['fan_normalized'] = (sports['fans'] - sports['fan_avg']) / sports['fan_stdev']
pd.set_option('precision', 2)
sports
```

You'll note that some functions take an optional argument, as is the case of std(), and the argument is supplied with in the parentheses:

<b>sports['fan_stdev'] = sports['fans'].std(ddof=0)</b>

This instructs the function to use zero "degrees of freedom", which is appropriate when considering a whole population.  The default is 1, which is typical when considering a sample.

You'll also note that we have a formatting expression that makes a setting to Pandas.   This statement:

<b>pd.set_option('precision', 2)</b>

...sets the display of floating point numbers just like the "Format Cells .. Number" operation within Excel.  The underlying values are not changed, just how they are presented.   If you want a permanent change you can use a function like round() or change the type.

Other built-in functions inclued median(), mode(), min(), max(), rank(), and kurtosis().

```python
#Here's one way to set the display precision
pd.set_option('precision', 4)

#We could change the underlying number with round()
sports['fan_avg_round'] = sports['fan_avg'].round(1)

#We could also change the variable type
sports['fan_avg_int'] = sports['fan_avg'].astype(int)

sports
```

### Functions From Afar

Because Pandas is built on top of other Python libraries, you have easy access to all sorts of other functionality.   If fact, if you dream up something you want to do with your data there's a good chance someone has already done it and incorporated it into a well-tested and reliable module.  It's often possible to do a bit of research and avoid re-inventing a wheel.

Two direct antecedents to Pandas, SciPy and NumPy, provide a rich source of math and science functionality that are "plug-compatible", available with a simple import.

To demonstrate, we'll play with some trig functions.  In the process, we'll build a new DataFrame from the ground up.  We're going to replicate 700+ rows of Excel spreadsheet with a few lines of code to accomplish this:

```python
#Some file names
support_dir = "support_files"
img_form_fn = os.path.join(support_dir, "trig_form.jpg")
img_data_fn = os.path.join(support_dir, "trig_data.jpg")

#Display Excel snippets
display_wide([img_form_fn, img_data_fn], spacing =2 ) 
```

You'll note that things are nearly-identical - except that you have to specify the equations only once.   You may also appreciate the flexibility the code-based approach provides.  This calculates sin and cos for two "revolutions" (720 degrees) for every degree.   To change the number of revolutions, calculate only 10 degree increments, etc. you need to change only one value - no cutting, pasting, copying, or deleting needed.

```python
#Import the NumPy library
import numpy as np
import pandas as pd

#Create data for the angles around a circle in increments
deg = [angle/1 for angle in range(0, 721, 1)]  # a list comprehension

#Make a DataFrame using the angles its first column
df = pd.DataFrame(data = deg, columns=['deg'])

#Make another column for radians (that's how NumPy thinks about angles).
df['rad'] = df['deg'].apply(np.math.radians)

#Now create a few trig columns
df['sin'] = np.sin(df['rad'])
df['cos'] = np.cos(df['rad'])

#Create a row index column
df.set_index('deg', inplace=True)
```

### Sorting Data

With Pandas there's no wondering "Did I sort that first?" because this is just another operation using a built-in function.   In Excel, you typically use the GUI's "Data .. Sort" dialog to choose columns, the sort hierarchy, and whether its ascending or descending.  In Pandas, you can make all these choices by providing parameters.

Here's how Pandas implents the sort capability:

- You can provide a list of the columns, in order.  The first is the primary sort column; ties are broken with the second; if there are still ties, they are broken with the third, and so on.  You can name the column list, if you want.  Something like:

<b>sort_columns = ['sin', 'cos']</b>

- To specify whether they are sorted in ascending order or not you can provide boolean values (the same number of values as you have columns to sort).  So you can go:

<b>sort_ascending = [True, False]</b>

- If you care where the "NaN" values go, you can also specify that they're 'first' or 'last'.  You can go something like:

<b>na_position = 'first'</b>

Putting it all together, here's how you might sort the trig DataFrame:

```python
sort_columns = ['sin', 'rad']
sort_ascending = [False, True]
na_position = 'first'
sorted_df = df.sort_values(by=sort_columns, 
                           ascending=sort_ascending, 
                           na_position=na_position)

#Switch the index to a zero-based integer value
sorted_df.reset_index(inplace=True)


##Graphing Boilerplate
#This is just display stuff and safe to ignore.
to_plot_0 = sorted_df.loc[:,'sin']
image_path_0 = os.path.join( "tmp", imgid)
plot_image_save(to_plot_0, image_path_0, 3, 3)
#
to_plot_1 = sorted_df.loc[:20, 'sin']
image_path_1 = os.path.join( "tmp", "_" + imgid)
plot_image_save(to_plot_1, image_path_1, 3, 3)
display_wide([sorted_df.iloc[:5,:] , image_path, image_path_1], 
             ["Data Sorted By Sin Then Cos", "Sin (2 Rotations)", "Sin 20 Rows"])
```

All you really need to specify is the column(s) by which you want to sort.  By default, it will put the NaN values at the end and sort ascending.

Note that we used the <b>iloc</b> method to pick the data to display.  <b>iloc</b> is short for "index locate" - instead of using the friendly names we may have given the rows and columns, Pandas uses the zero-based index that it retained.  As you can see it's used just like "locate" <b>loc</b>.

### Named Ranges and Views

Ranges provide a breath of sanity within complicated spreadsheets.  The allow sub-regions to be identified and addressed independently.

Pandas' take on the ranges is all about memory.  It's possible to create new objects based on an origional that share memory with the original.   That is - anything you do to the new objects (called "views") is reflected in the original.

Here's a view based on the top few rows of the trig DataFrame.  We can use the function <b>numpy.shares_memory()</b> to verify that we have a view and not just a copy:

```python
top = df.iloc[:3,:]

#Check if this object shares memory and whether it's a view
print("Do 'df' and 'top' share memory?  {}".format(np.shares_memory(top, df)))
print("Is 'top' a view?   {}".format(top._is_view))

display_wide([df.iloc[:5,:] , top], ["Original", "View"], spacing=3)
```

With a bite-sized chunk of the larger object in hand we can make some arbitrary changes that affect just a few rows:

```python
#The default is to generate a warning for the unwary.  Turn it off.
pd.set_option('mode.chained_assignment', None)

#Alter some values on the view
top['rad'] = top['rad'] + 100
top['sin'] = top['sin'].apply(lambda x: x * 100 + 777)
top['cos'] = top['cos'] * 1000.

#Remember to turn the warning back on!
pd.set_option('mode.chained_assignment', 'raise')

display_wide([df.iloc[:5,:] , top], ["Original", "View"], spacing=3)
```

By inspection you can observe that the operations on the object "top" affected only the first three rows of the original object.  The rest of the rows escaped unscathed.

You want to be really careful when operating on data views like this.  Pandas will make all sorts of racket when you try it, but it's easy enough to grow complaciant around silencing the warnings.  Since these operations can have far-reaching "stealth" effects beyond the visibility of locally-scoped routines they can easily go undetected.

More typically you'll be working on copies of data objects, anyway.   You automatically get a copy with just about any operation except a slice - that goes for queries, column-wise operations (<b>apply()</b> operates on individual "cells"), and row-wise operations.

If you're ever unsure, either use the <b>shares_memory()</b> against both candidate objects method or ask for the <b>_is_view</b> property of the suspected copy.

```python
# A sure-fire way to get a clean copy
top_copy = df.iloc[:3,:].copy()  

#Check if the copy shares memory and whether it's a view
print("Do 'df' and 'top_copy' share memory?  {}".format(np.shares_memory(top_copy, df)))
print("Is 'top_copy' a view?   {}".format(top_copy._is_view))
```

### Lookups

In Excel, it's common to add data to a calculation by finding it in some sort of lookup table.  You might use INDEX() if you know the row and column numbers.  You might use OFFSET() if you know wher e the data lives relative to some specific place in the spreadsheet.

More typically, value-based functions such as HLOOKUP() and VLOOKUP() are used to match up rows and columns based on "what" rather than "where".

We'll begin by showing you a "direct port" of the Excel functionality then demonstrate a few of the many upgrades available.  In the proces, you'll learn some more of Panda's and Excels cababilities.

<u>INDEX() and OFFSET()</u>

The translation of INDEX() is pretty straightforward - the only twist is that Excel uses 1 as the starting point and Python uses 0, but that's not a big deal.  To demonstrate we'll bring in the sports data, allowing Pandas to set its native index values for both rows and columns:

```python
#Excel = INDEX(sports, 3, 2)

sports = pd.read_csv(csv_file_name, header=None)
row = 3
col = 2

#Use iloc() to mimic INDEX()
value = sports.iloc[row, col]

display_wide([sports ,  value],  spacing=2)
```

The translation of OFFSET() is equally straightforward.  From some reference point, you can simply increment the row and column indices.  This is just like working with INDEX(), but adding the offset values.

```python
#Excel = OFFSET(INDEX(sports, 0, 2), 2, 1)
reference_row = 0;  row_offset = 2
reference_col = 1;  col_offset = 1

#Use iloc() to mimic OFFSET()
value = sports.iloc[reference_row + row_offset,
                    reference_col + col_offset]

display_wide([sports ,  value],  spacing=2)
```

<u>VLOOKUP() and HLOOKUP()</u>

A more flexible (and human-friendly) approach is to identify data based on something besides its absolute or relative postion.  These might change as our project evolves, anyway.

In Excel, VLOOKUP() is used to find values in the right row and HLOOKUP() is used to find values in the right column.  They work identically so we'll just focus on VLOOKUP() here.

The VLOOKUP() function will search the first column of a block of data for the first occurrance of a given value.  This identifies the row.  Within that row, it will use the provided offset to find the right value.  If you're feeling brave, you can also ask it to guess the right row if the data doesn't match.  The syntax is something like:

=VLOOKUP ( what_to_find, range_to_search, col_to_return, FALSE_for_avoid_guesswork)

With Pandas, we can use the <b>loc[ate]</b> method to search the index column for the right value, then the right column.  To demonstrate, we'll let Pandas set the row index to the name of the sport.  We'll have to calculate the numeric value of the column index here (but that's only to demonstrate the direct translation from Excel).

```python
#Excel:  =VLOOKUP('baseball', sports, 1, FALSE)

sports = pd.read_csv(csv_file_name, index_col='sport')

#Parameters for VLOOKUP
what_to_find = 'baseball'
col_to_return = 1
range_to_search = sports

#Find real column name based on column provided
col_name = sports.columns[col_to_return]

#Use loc() to mimic VLOOKUP()
value = sports.loc[what_to_find, col_name]

display_wide([sports ,  value],  spacing=2)
```

### Lookup Upgrades

We've demonstrated that you can direcly map your existing work with Excel to Panda.  In the process, you might consider deviating somewhat to take advantage of Pandas' enhanced capabilities.   For instance you can:

-  Find values in a data table without rearranging columns or adjusting indices.
-  Find multiple matches, if they exist
-  Simultaneously search multiple columns for more complex queries
-  Return intact rows of data instead of single values.



<u>Find data without rearranging columns</u>

In Excel, you need a stable and known arrangemnet of data to use tools like VLOOKUP().   For instance, the information you're trying to match has to be in the first column.  This can be tedious and trigger error-prone, mechanical operations.

With Pandas you can take a more direct approach.   Let's say you wanted the number of fans attending an event that lasts 30 minutes, but you don't know what sport that would be.  

Here's how you might do it:

```python
sports = pd.read_csv(csv_file_name, index_col='sport')

#Grab the column you care about and apply a test 
test_results = sports['duration'] == 30

#Use the results as a filter to get matching row(s)
row = sports[test_results]

#From within matching row(s) find the right column
value = row['fans']

display_wide([sports ,    test_results,   row,                       value],
             ["Raw Data", "Test Results", "Use Results to get Row", "Find Value in Row"],
               spacing=1)
```

There's much to go over here ... so let's break it down.

<b>test_results = sports['duration'] == 30</b>

<b>sports['duration']</b> returns the 'duration' column from the original DataFrame (a reference to it, really).   When we apply the test <b> == 30</b> we are asking for a True/False evaluation.   Like the other operations we've seen, this evaluation is automatically applied to each row within the column.  This is shown as "Test Results".

A new column object, called a pandas.Series, is automatically created as we perform the test.  The Series object inherits the row index of the original DataFrame.  We can see that the 'wrestling' row has been evaluated as True and the others False.

We can use this new column as sort of a row filter against the original data.  Because the two objects share index values, we can use the new column to choose which row(s) we want.  These two expressions are equivalent, and are shown as "Use Results to get Row".

<b>row = sports[test_results]</b>
<b>row = sports.loc[test_results, :]</b>

Now, having selected the right row, we can then specify which column we want, shown as "Find Value in Row":

<b>value = row['fans']</b>

Here are all the steps, rolled together in one elegant equation:

```python
#Average attendance for sports lasting a half hour

#                   ----- row selector ----     - cols -
value = sports.loc[ sports['duration'] == 30  ,  'fans']

display_wide(value)
```

You might regard it as weird that we can "gang up" Pandas expressions like this.  We have a column, then a test, then a data selection operation - and we did it all at once.   But if you think about it, this is no different than calculating an Excel cell based on the value of several others - even when the others are, themselves, calculated using other cells.  We have flexibility in this regard because Excel cells are, by and large, compatible with one another.

Under the hood, most operations on Pandas objects produce new Pandas objects.   Both the new and existing objects have the same methods and attributes.  They may even have the same index values.  You have to check, but more often than not they are completely compatible with one another.   As a result you can take the results from one operation and apply them to another operation ... and another ... and another .. all in a single expression.

At any point you can "freeze" the process and produce an permanent version intermediate object by giving it a name.  We did that with this equation:

<b>test_results = sports['duration'] == 30</b>

... but that was a choice - in this case for clarity of expression - not a requirement.  It's very common to use these "componded expressions" for clarity of <u>programming</u>.  Everything you need is compactly settled in one place so you can grasp it without reviewing multiple lines of code.

With your own work, you can use a compact or verbose style as you see fit.   You may be more comfortable with creating multiple objects so you know just what you're dealing with.  That's completely OK - especially when your data objects isn't a big enough to be factor in performance or turn-around time.
